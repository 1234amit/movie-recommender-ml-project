Step 2 — Import Libraries

```python
import pandas as pd
import numpy as np
```

Step 3 — Load Dataset

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

Step 4 — Select Important Columns

```python
movies = movies[['id','original_title','genres','keywords','overview']]
```

```python
#cheching missing values
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
#Step 6 — Combine Text Features
movies['tags'] = movies['genres'] + " " + movies['keywords'] + " " + movies['overview']
```

```python
#Step 7 — Convert Text → Numbers (Vectorization)
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()
```

Step 8 — Calculate Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

Step 9 — Create Recommendation Function

```python
# def recommend(movie):
#     movie_index = movies[movies['original_title'] == movie].index[0]
#     distances = similarity[movie_index]
#     movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
#
#     for i in movie_list:
#         print(movies.iloc[i[0]].original_title)

def recommend(movie):

    if movie not in movies['original_title'].values:
        print("Movie not found. Please check the spelling.")
        return

    movie_index = movies[movies['original_title'] == movie].index[0]

    distances = similarity[movie_index]

    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movie_list:
        print(movies.iloc[i[0]].original_title)

```

Test

```python
recommend("Iron Man")
```
